```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('MOVIE_IMDB.csv')
```

```python
df
```

```python
clm=df.columns
clm_nan_num=[]
for i in clm:
    clm_nan_num.append(df[i].isna().sum())#number of nan in each column
clm_nan_perc=((np.array(clm_nan_num)/5043)*100).round(2)#percent of nan in each column
clm_num_perc= list(zip(clm_nan_num,clm_nan_perc))
pd.DataFrame(clm_num_perc,index=clm,columns=['Number_of_NaNs','Percent_of_NaNs'])
```

```python
row_nan_num=[]
for i in range(0,5043):
    row_nan_num.append(df.loc[[i]].isna().sum().sum())#number of nan in each row
row_nan_perc=((np.array(row_nan_num)/28)*100).round(2)#percent of nan in each row
row_num_perc= list(zip(row_nan_num,row_nan_perc))
pd.DataFrame(row_num_perc,columns=['Number_of_NaNs','Percent_of_NaNs'])
```

```python
#adding profit column
profit=df_1['gross']-df_1['budget']
df_1['profit']=profit
```

```python
pos_pof=df_1.loc[df_1['profit']>0,'profit']
pos_pof_bud=df_1.loc[df_1['profit']>0,'budget']

plt.rcParams['figure.figsize'] = [16, 8]
plt.plot(pos_pof_bud/10**8,pos_pof/10**8,'r.')
plt.grid()
plt.xlabel('$Budget (x10^8)$')
plt.ylabel('$Profit (x10^8)$')
plt.title('Profit-Budget Plot')
```

```python
#geners
all_unique_geners=np.array(df_1['genres'].value_counts().index)
print(all_unique_geners.size,all_unique_geners)
```

```python
# number of movie in each gener
num_mov_geners=np.array(df_1['genres'].value_counts().values)
# print(num_mov_geners.size,num_mov_geners)
pd.DataFrame(num_mov_geners,index=all_unique_geners,columns=['Number_of_movie'])
```

```python
#average of gross in each genre
sum_gros=[]
for i in all_unique_geners:
     sum_gros.append((df_1[df_1['genres']==i]['gross'].sum()))

mean_gross_of_genr=sum_gros/num_mov_geners
pd.DataFrame(mean_gross_of_genr,index=all_unique_geners,columns=['average of gross in each genre']).head(20)
```

relation of gross and people's intrest (the last part)

```python
# relation of gross and people's intrest (the last part)
# gross-IMDB_Score Plot
plt.rcParams['figure.figsize'] = [16, 8]
plt.plot(df_1['imdb_score'],df_1['gross']/10**8,'r.')
plt.grid()
plt.xlabel('IMDB_Score')
plt.ylabel('$gross (x10^8)$')
plt.title('gross-IMDB_Score Plot')
```

```python
uniq_score=np.array(df_1['imdb_score'].value_counts().index)
num_mov_in_esch_uni_score=np.array(df_1['imdb_score'].value_counts())

sum_gros_score=[]
for i in uniq_score:
     sum_gros_score.append((df_1[df_1['imdb_score']==i]['gross'].sum()))
mean_gross_of_score=sum_gros_score/num_mov_in_esch_uni_score

plt.rcParams['figure.figsize'] = [16, 8]
plt.plot(uniq_score,mean_gross_of_score/10**8,'ro')
plt.grid()
plt.xlabel('IMDB_Score')
plt.ylabel('$ average_of_gross (x10^8)$')
plt.title('average of gross-IMDB_Score Plot')
```

```python
# gross - movie_facebook_likes Plot

non_zero_likes=df_1.loc[(df_1['movie_facebook_likes']>0) & (df_1['movie_facebook_likes']<10**5),'movie_facebook_likes']
gross_non_zero_likes=df_1.loc[(df_1['movie_facebook_likes']>0) & (df_1['movie_facebook_likes']<10**5),'gross']

plt.rcParams['figure.figsize'] = [16, 8]
plt.plot(non_zero_likes,gross_non_zero_likes/10**8,'r.')
plt.grid()
plt.xlabel('Number of facebook likes')
plt.ylabel('$gross (x10^8)$')
plt.title('gross-Number of facebook likes Plot')
```
